# Packages

```python
# data handling
import numpy as np
import pandas as pd

# machine learning
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
```

# Read in Data

```python
genre_list = ['indie_alt',
              'alternative',
              'blues',
              'hiphop',
              'metal',
              'pop',
              'rock']

indie_alt = pd.read_csv(genre_list[0] + '_music_data.csv')
alt = pd.read_csv(genre_list[1] + '_music_data.csv')
blues = pd.read_csv(genre_list[2] + '_music_data.csv')
hiphop = pd.read_csv(genre_list[3] + '_music_data.csv')
metal = pd.read_csv(genre_list[4] + '_music_data.csv')
pop = pd.read_csv(genre_list[5] + '_music_data.csv')
rock = pd.read_csv(genre_list[6] + '_music_data.csv')
```

```python
# now we can make a separate table which contains all the genres as variables with their corresponding tables
genres = [indie_alt,
         alt,
         blues,
         hiphop,
         metal,
         pop,
         rock]

# this prints the names of the genres themselves rather than their tables
print('Genre List:', genre_list) 
```

```python
# we can access the data by calling the variables directly
hiphop.head()
```

```python
# or we can access a table by calling it from our 'genres' list
genres[4].head()
```

# Motivation

The goal for this project is to classify data using a machine learning classification algorithm. Since I will be working with multiple datasets, much of the work will come with pre-processing all the data and combinining the data in such a way that the title of each individual dataset is a value in the larger, combined dataset, namely this value is used to compute the accuracy of the classification algorithm.

# Pre-Processing Data

### Adding the Genre for Each Table

It will be easier to work with the data once it is all in one table, however, the most immediate concern is being able to discern the different overarching genres (which is unique from the 'Genres' column present in all the data). We want to work with a single classification.

```python
for table,name in zip(genres,genre_list):
    table['genre'] = name
```

```python
hiphop.head()
```

As we can see, all the tables will now have their corresponding genre. Thus, we can now combine the data. Luckily, this should be straightforward since we have all our tables in one list. 

### Combining All the Tables

```python
# this combines all the tables
df = pd.concat(genres)

# we also want to shuffle around the rows of our new table, and make sure to not preserve the old indexes
df = df.sample(frac=1).reset_index(drop=True)

print('Table size:', df.size)
```

### Finding Duplicate Songs

```python
# we have to convert the values in our 'Track Name' column to string
df['Track Name'] = df['Track Name'].astype(str)
```

```python
# finds the duplicate songs in the dataframe
df.drop_duplicates(subset='Track Name', keep=False, inplace=True)
```

```python
print("Updated number of songs:", df['Track Name'].size)
```

### Dropping Redundant Columns

```python
# creating a separate list to drop allows us to better visualize which columns are being dropped from the data
drop_these_columns = ['Artist Name',
                     'Track Name',
                     'Popularity',
                     'Genres',
                     'Playlist',
                     'energy',
                     'id',
                     'uri',
                     'track_href',
                     'analysis_url',
                     'time_signature']

# drops the redundant columns
df.drop(drop_these_columns, inplace=True, axis=1)
```

```python
# updated dataframe without redundant columns
df.head()
```

At this point, we've removed all duplicate songs, as well as removed all the columns that aren't useful in our classification. We don't have to worry about null values in the dataset, as prior work with the data showed that the data did not have any.

# Implementing the Classification Algorithm

### Splitting the Data

```python
# setting the parameters and target variable
x = df[['danceability',
            'key',
            'loudness',
            'mode',
            'speechiness',
            'acousticness',
            'instrumentalness',
            'liveness',
            'valence',
            'tempo',
            'duration_ms']]

y = df['genre']

# spliting the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

### Building our K-Nearest Neighbors Model

```python
# initiating the model
knn_model = KNeighborsClassifier(n_neighbors=50)

# fitting the data to the model
knn_model.fit(x_train, y_train)

# making the classification prediction
knn_prediction = knn_model.predict(x_test)
```

### Evaluating the Accuracy of the Model

```python
print(classification_report(knn_prediction, y_test))
```

Looking at the classification report, we can see that the model performed rather terribly for classifying songs as 'alternative', bad for most of the other genres, and somewhat decently for the 'rock' genre (where the classifier was correct in its prediction of a song being of the 'rock' genre 76% of the time).

From here, we can look at pitting the 'rock' genre against one of the other genres, perhaps 'hiphop', and look to create a binary classifier.

# Building a Binary Classifier

### Creating a Binary Dataset

```python
df['genre'] = df['genre'].astype(str)

# creting new datasets of both pop and rock songs
pop_df = df.loc[df['genre'] == 'pop']
rock_df = df.loc[df['genre'] == 'rock']

# displaying the number of songs in the dataset
print("Number of pop songs:", pop_df.loc[pop_df['genre'] == 'pop'].size)
print("Number of rock songs:", rock_df.loc[rock_df['genre'] == 'rock'].size)
```

```python
# reducing the number of songs so that each dataset is equal
reduced_pop_df = pop_df.iloc[:20000]
reduced_rock_df = rock_df.iloc[:20000]

# combining our two reduced datasets
bin_df = pd.concat([reduced_pop_df, reduced_rock_df])
```

```python
#shuffling the binary dataset
bin_df = bin_df.sample(frac=1).reset_index(drop=True)

# our new binary dataset below
bin_df.head()
```

### Implementing Another K-Nearest Neighbor Model

```python
# setting the parameters and target variable
bx = bin_df[['loudness',
            'mode',
            'speechiness',
            'acousticness',
            'instrumentalness',
            'liveness',
            'valence',
            'tempo']]

by = bin_df['genre']

# spliting the data into training and testing sets
bx_train, bx_test, by_train, by_test = train_test_split(bx, by, test_size=0.3)
```

```python
# initiating the model
knn_bin_model = KNeighborsClassifier(n_neighbors=15)

# fitting the data to the model
knn_bin_model.fit(bx_train, by_train)

# making the classification prediction
knn_bin_prediction = knn_bin_model.predict(bx_test)
```

### Binary Classification Evaluation

```python
print(classification_report(knn_bin_prediction, by_test))
```

```python
print("Accuracy of the Model:", accuracy_score(knn_bin_prediction, by_test))
```

# Analysis of the Results

We built two different classifiers here. One a multiclass classifier and a binary classifier. We can see that the binary classifier naturally worked better than the multiclass one did. However, the model seemed to perform best on classifing songs as 'rock'. A more in-depth analysis of the parameters might reveal why this is, namely, a certain parameter or list of parameters might be better at predicting the genre of a song as 'rock' simply because rock songs tend to have specific patterns. Regardless, the model is not great at predicting if a song is another genre.
